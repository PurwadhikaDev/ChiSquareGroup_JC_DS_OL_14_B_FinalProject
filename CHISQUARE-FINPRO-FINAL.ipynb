### Metrics Evaluation:

target:<br>
1 : Nasabah yang berpotensi membeli deposito berjangka<br>
0 : Nasabah yang tidak berpotensi membeli deposito berjangka

- TP = Memprediksi nasabah yang berpotensi membeli deposito berjangka
- TN = Memprediksi nasabah yang tidak akan membeli deposito berjangka
- FP = Salah memprediksi nasabah yang berpotensi membeli deposito berjangka, sebenarnya tidak akan membeli deposito berjangka. Sehingga perlu mengeluarkan biaya dan waktu yang lebih kepada nasabah yang sebenarnya tidak akan membeli deposito berjangka
- FN = Salah memprediksi nasabah yang tidak akan membeli deposito berjangka, sebenarnya berpotensi akan membeli deposito berjangka. Sehingga kehilangan calon nasabah yang berpotensi akan mengambil membeli deposito berjangka (kehilangan profit)


Dalam memprediksi apakah seorang klien akan berlangganan deposito berjangka atau tidak, pemilihan metrik evaluasi yang tepat sangat penting untuk memastikan keberhasilan model dalam mendukung keputusan bisnis. Berikut adalah risiko dari masing-masing klasifikasi salah, serta metrik evaluasi yang digunakan:

- *False Positive (FP)*: Klien yang diprediksi akan tertarik berlangganan, tetapi kenyataannya tidak. Kesalahan ini berisiko pada biaya kampanye pemasaran yang sia-sia, karena upaya dan sumber daya dikeluarkan untuk klien yang tidak tertarik.
  - *Alasan*: FP berdampak pada biaya kampanye yang signifikan, sehingga metrik Precision lebih tepat untuk meminimalkan kesalahan ini. Dengan fokus pada precision, kita berusaha memastikan bahwa sumber daya hanya diarahkan pada klien yang benar-benar tertarik, sehingga mengurangi pemborosan biaya kampanye.

- *False Negative (FN)*: Klien yang diprediksi tidak tertarik, tetapi sebenarnya berpotensi untuk berlangganan. Kesalahan ini dapat menyebabkan peluang pemasaran yang hilang.
  - *Alasan*: FN berisiko pada hilangnya kesempatan, namun dalam konteks ini risiko kehilangan pelanggan potensial lebih kecil dibandingkan biaya yang dikeluarkan untuk kampanye yang salah sasaran. Oleh karena itu, precision lebih diutamakan.

*Conclusion*:
Dalam kampanye pemasaran ini, Precision lebih diutamakan karena risiko finansial dari false positives (FP) lebih signifikan dibandingkan risiko false negatives (FN).

### Note:
- The dataset is **imbalanced**; the majority of clients did not subscribe to a term deposit.

Oleh karena itu, metode penanganan data imbalance seperti *SMOTE* digunakan untuk meningkatkan performa model.

### Pentingnya Mengatasi Ketidakseimbangan:
- *False Negatives (FN):* Dalam kasus ketidakseimbangan data, model dapat lebih sering mengklasifikasikan kelas "Yes" sebagai "No," yang berarti peluang potensial untuk penjualan deposito berjangka bisa terlewatkan.
- *Solusi:* Teknik *SMOTE (Synthetic Minority Over-sampling Technique)* dapat digunakan untuk menambah jumlah contoh pada kelas minoritas. Dengan cara ini, model dapat dilatih dengan distribusi yang lebih seimbang dan diharapkan menghasilkan prediksi yang lebih akurat untuk kedua kelas.

Dataset ini menunjukan bahwa data imbalance

### 4. Preprocessing Data

Preprocessing merupakan tahap penting dalam menyiapkan data sebelum dilakukan pelatihan model. Pada langkah ini, beberapa teknik *encoding* dan *scaling* digunakan:

- *RobustScaler*: Digunakan untuk fitur numerik seperti age, campaign, pdays, dan previous untuk menangani **outlier* dalam data.
- *OneHotEncoder*: Digunakan untuk fitur kategorikal seperti job, marital, contact, loan, dan housing yang tidak memiliki urutan. Teknik ini mengubah kategori menjadi representasi biner.
- *OrdinalEncoder*: Digunakan untuk fitur kategorikal ordinal seperti education, month, dan day_of_week, yang memiliki urutan logis.

Encoding ini memastikan bahwa semua fitur baik numerik maupun kategorikal dapat digunakan secara efektif oleh model machine learning.

```python
# Create Preprocesing
numerical_features = ['age', 'campaign', 'pdays', 'previous']
categorical_features_onehot = ['job', 'marital', 'poutcome']
categorical_features_label = ['housing', 'loan', 'contact']
categorical_features_mapping = [
            {'col': 'housing', 'mapping': {'no': 0, 'yes': 1}},
            {'col': 'loan', 'mapping': {'no': 0, 'yes': 1}},
            {'col': 'contact', 'mapping': {'telephone': 0, 'cellular': 1}}
            ]

ordinal_features = ['education', 'month', 'day_of_week']

# Mapping untuk ordinal encoding (jika ada urutan)
ordinal_mapping = [
    {'col': 'education', 'mapping': {'basic.4y': 1, 'basic.6y': 2, 'basic.9y': 3, 
                                     'high.school': 4, 'professional.course': 5, 
                                     'university.degree': 6, 'illiterate': 7}},
    {'col': 'month', 'mapping': {'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 
                                 'jun': 6, 'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 
                                 'nov': 11, 'dec': 12}},
    {'col': 'day_of_week', 'mapping': {'mon': 1, 'tue': 2, 'wed': 3, 'thu': 4, 'fri': 5}}
    ]

# Buat preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', RobustScaler(), numerical_features),  # Scaler numerikal
        ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'), categorical_features_onehot),  # One-Hot Encoder
        ('label', ce.OrdinalEncoder(mapping=categorical_features_mapping), categorical_features_label),  # Label Encoding via OrdinalEncoder
        ('ordinal', ce.OrdinalEncoder(mapping=ordinal_mapping), ordinal_features)  # Ordinal Encoding
    ], 
    remainder='passthrough'  # Untuk fitur lainnya yang tidak didefinisikan
)

preprocessor

```

## Methodology (Modeling/Analysis)

```python
#Cross validation
models = {
    'LogReg': LogisticRegression(),
    'DecisionTree': DecisionTreeClassifier(),
    'KNeighbors': KNeighborsClassifier(),
    'RandomForest': RandomForestClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'XGBoost': XGBClassifier(),
    'GBooost': GradientBoostingClassifier(),
    'LightGBM': lgb.LGBMClassifier()
}


results = []

# Loop through the models
for i in models:
    skfold = StratifiedKFold(n_splits=5)
    
    # Create an estimator pipeline
    estimator = Pipeline(steps=[
        ('preprocess', preprocessor), # Apply preprocessing
        ('model', models[i])           # Model step
    ])
    
    # Perform cross-validation
    model_cv = cross_val_score(estimator, xtrain, ytrain, cv=skfold, scoring='precision')
    
    # Append the results
    results.append({
        'Model': i,
        'Precision Scores': model_cv,
        'Mean Precision': model_cv.mean(),
        'Standard Deviation': model_cv.std()
    })

# Convert results to a DataFrame and sort by Mean Precision
df_results = pd.DataFrame(results).sort_values('Mean Precision', ascending=False)
df_results
```

Model Logistic Regression memiliki nilai rata-rata Precision evaluasi metrics yang terbaik, disusul dengan model  GradientBoosting dan AdaBoost selain memiliki nilai evaluasi matrics yang baik, keduanya memiliki standar devisiasi yang tidak besar menandakan keduanya memiliki nilai evaluasi metrics yang stabil.
Maka dari itu, Ketiga model ini yang akan dilakukan test dengan data test.

```python
#test data benchmark
# Apply preprocessing to xtrain and xtest
xtrain_transformed = preprocessor.fit_transform(xtrain)
xtest_transformed = preprocessor.transform(xtest)

results = []

# Loop through each model
for name, model in models.items():
    # Fit the model
    model.fit(xtrain_transformed, ytrain)
    
    # Predict on the test set
    y_pred = model.predict(xtest_transformed)
    
    # Calculate precision score
    precision = precision_score(ytest, y_pred)
    
    # Append the results in the list
    results.append({
        'Model': name,
        'Precision': precision
    })
    
    # Print the classification report for each model
    print(f"Classification Report for {name}:\n")
    print(classification_report(ytest, y_pred))
    print("="*60)

# Convert the results into a DataFrame for easy visualization
results_df = pd.DataFrame(results).sort_values('Precision', ascending=False)
results_df
```

```python
# Test Oversampling with K-Fold Cross Validation
models = {
    'LogReg': LogisticRegression(max_iter=1000),
    'DecisionTree': DecisionTreeClassifier(),
    'KNeighbors': KNeighborsClassifier(),
    'RandomForest': RandomForestClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    'GBooost': GradientBoostingClassifier(),
    'LightGBM': lgb.LGBMClassifier()
}

# Create an empty list to store the results
results = []

# Apply preprocessing to xtrain and xtest
xtrain_transformed = preprocessor.fit_transform(xtrain)
xtest_transformed = preprocessor.transform(xtest)

# Oversampling using SMOTE
smote = SMOTE(random_state=42)
xtrain_resampled, ytrain_resampled = smote.fit_resample(xtrain_transformed, ytrain)

# Loop through each model and perform benchmarking
for name, model in models.items():
    # ----------------------
    # Without oversampling
    # ----------------------
    # Fit the model
    model.fit(xtrain_transformed, ytrain)
    
    # Predict on the test set
    y_pred_without_oversampling = model.predict(xtest_transformed)
    
    # Calculate precision, accuracy, and F1-score
    precision_without = precision_score(ytest, y_pred_without_oversampling)
       
    # ----------------------
    # With oversampling (SMOTE)
    # ----------------------
    # Fit the model on the resampled training set
    model.fit(xtrain_resampled, ytrain_resampled)
    
    # Predict on the test set
    y_pred_with_oversampling = model.predict(xtest_transformed)
    
    # Calculate precision, accuracy, and F1-score
    precision_with = precision_score(ytest, y_pred_with_oversampling)
        
    # Append the results to the list
    results.append({
        'Model': name,
        'Precision Without SMOTE': precision_without,
        'Precision With SMOTE': precision_with
            })

# Convert the results into a DataFrame for easy visualization
results_df = pd.DataFrame(results).sort_values(by='Precision With SMOTE', ascending=False)
results_df
```

## Model Limitation

- batasan usia

Model ini dilatih menggunakan data dengan rentang usia nasabah antara 17 hingga 100 tahun.

- anomali unknown

Data "unknown" yang terlalu banyak bisa mengganggu proses pelatihan model karena memberikan informasi yang ambigu atau tidak jelas. Model mungkin salah mengasumsikan bahwa "unknown" adalah kategori yang berarti, padahal itu menunjukkan data yang hilang atau tidak tersedia.

- metric hanya precission

Fokus hanya pada precision tanpa memperhitungkan metric lain, seperti F1-score, dapat menyebabkan model cenderung mengabaikan prediksi untuk kelas minoritas (misalnya, model tidak memprediksi kelas positif sama sekali untuk mendapatkan precision tinggi).

- data imbalance
terhadap ketidakseimbangan data yang telah dilakukan oversampling tetap menghasilkan data yang belum optimal
